```python
df3 = df2.copy()
df3.drop(['id','date', 'bedrooms', 'sqft_living','floors', 'view', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'lat', 'long', 'sqft_lot15', 'sqft_living15'], axis = 1, inplace = True)
df3['tennis_within_5'] = [1,2,5,2,0]
df3['golf_within_5'] = [2,1,2,1,0]
df3['idx_west_to_east'] = [ 4, 3,1 ,2 , 5]
df3.sort_values(by= "idx_west_to_east", inplace = True)
df3['Location'] = ['Shoreline','Downtown','Bellevue','Redmond','StateForest']
df3['Price_per_sqf'] = df3.price / df3.sqft_lot
df3['weighted_criteria'] = 0.1 * df3.price/df3.price.max() + 0.5 * df3.sqft_lot/df3.sqft_lot.max() + 0.2 * df3.tennis_within_5/df3.tennis_within_5.max() + 0.2 * df3.golf_within_5/df3.golf_within_5.max()
df3['weighted_criteria_%'] = df3.weighted_criteria * 100
df3.head()
```

```python
import matplotlib.pyplot as plt


#f1 = df3.plot(kind = 'bar',x='Location',y='price') # das sollte ein ganz normaler pandas plot sein!

plt.figure()
plt.figure(figsize=(14,8))  # figsize=(float, float) : Width, height in inches
plt.suptitle("Houses meeting described criteria, 4+ bathrooms", fontsize=20) # add a super title to the container
plt.style.use('fast')

#['Solarize_Light2', '_classic_test_patch', '_mpl-gallery', '_mpl-gallery-nogrid', 'bmh', 'classic',\
#  'dark_background', 'fast', 'fivethirtyeight', 'ggplot', 'grayscale', 'seaborn', 'seaborn-bright', \
# 'seaborn-colorblind', 'seaborn-dark', 'seaborn-dark-palette', 'seaborn-darkgrid', 'seaborn-deep', \
# 'seaborn-muted', 'seaborn-notebook', 'seaborn-paper', 'seaborn-pastel', 'seaborn-poster', \
# 'seaborn-talk', 'seaborn-ticks', 'seaborn-white', 'seaborn-whitegrid', 'tableau-colorblind10']

###########################
plt.subplot(2,2,1) # (rows, columns, panel number)
plt.title('Real estate (from west to east) vs. price')
plt.ylabel('Million Dollar')
plt.tick_params(
    axis = 'x',
    which = 'both',
    bottom = False
    )
plt.bar(df3['Location'], df3['price'] / 1000000 , align = "center");

#############################
plt.subplot(2,2,2)
plt.title('Real estate (from west to east)) vs. square feet of property')
plt.ylabel('Square feet x 1000')
plt.bar(df3['Location'], df3['sqft_lot'] / 1000 , align = "center");

##############################
plt.subplot (2,2,3)
plt.title('Real estate (from west to east) vs. Price per sqft')
plt.ylabel('Price per sqft')
plt.bar(df3['Location'], df3['Price_per_sqf'] , align = "center");

# ##############################
# plt.subplot (2,2,4)
# plt.title('Company recommendation for visiting real estates')
# plt.ylabel('% of weighted criteria')
# plt.bar(df3['Location'], df3['weighted_criteria_%'] , align = "center");

plt.subplot (2,2,4)
plt.title('Golf/Tennis court within 5 miles')

plt.ylabel('Number of courts')
width = 0.8
bottom = np.zeros(5)
dict_t_and_g = {
    "Tennis": np.array([5,2,2,1,0]),
    "Golf": np.array([2,1,1,2,0]),
}
for boolean, t_g in dict_t_and_g.items():
    p = plt.bar(df3.Location , t_g, width, label = boolean, bottom = bottom)
    bottom += t_g
plt.legend(loc = "upper right")
```
